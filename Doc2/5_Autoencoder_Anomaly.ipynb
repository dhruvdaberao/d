```python
# Practical 5: Autoencoder for Credit Card Anomaly Detection

# Step 1: Import Libraries
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# Step 2: Load and Preprocess Data
# (You can download the dataset: https://www.kaggle.com/mlg-ulb/creditcardfraud)
data = pd.read_csv('creditcard.csv')

# Separate features and labels
X = data.drop('Class', axis=1)
y = data['Class']

# Normalize feature data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Step 3: Define Autoencoder Architecture
input_dim = X_train.shape[1]
encoding_dim = 14  # latent space dimension

input_layer = Input(shape=(input_dim,))
encoder = Dense(encoding_dim, activation="relu")(input_layer)
decoder = Dense(input_dim, activation="sigmoid")(encoder)
autoencoder = Model(inputs=input_layer, outputs=decoder)
```

```python
# Step 4: Compile the Model
autoencoder.compile(optimizer='adam', loss='mse')
```

```python
# Step 5: Train the Model
history = autoencoder.fit(X_train, X_train,
                          epochs=20,
                          batch_size=64,
                          shuffle=True,
                          validation_data=(X_test, X_test),
                          verbose=1)
```

```python
# Step 6: Evaluate Model and Compute Reconstruction Error
reconstructions = autoencoder.predict(X_test)
mse = tf.keras.losses.mse(X_test, reconstructions)
error = mse.numpy()
```

```python
# Step 7: Visualize Reconstruction Error
plt.figure(figsize=(10, 5))
plt.hist(error, bins=50)
plt.title("Reconstruction Error Distribution")
plt.xlabel("Reconstruction Error")
plt.ylabel("Count")
plt.show()
```

```python
# Step 8: Detect Anomalies (Optional)
threshold = error.mean() + 3 * error.std()
anomalies = error > threshold
print("Detected anomalies:", anomalies.sum())
print("Actual fraud cases in test set:", y_test.sum())
```
